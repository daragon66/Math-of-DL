```python
def subsample(x_in):
    height, width = x_in.shape
    x_out = np.zeros((int(np.ceil(height / 2)), int(np.ceil(width / 2))))

    for i in range(0, height, 2):
        for j in range(0, width, 2):
            x_out[i // 2, j // 2] = x_in[i, j]

    return x_out
```

```python
# Now let's try max-pooling
def maxpool(x_in):
  x_out = np.zeros(( int(np.floor(x_in.shape[0]/2)), int(np.floor(x_in.shape[1]/2)) ))
  # TO DO -- write the maxpool routine
  # Replace this line
  x_out = x_out

  return x_out
```

```python
def maxpool(x_in):
    height, width = x_in.shape
    new_height = height // 2
    new_width = width // 2
    x_out = np.zeros((new_height, new_width))

    for i in range(new_height):
        for j in range(new_width):
            # Extract a 2x2 patch from the input
            patch = x_in[2*i:2*i+2, 2*j:2*j+2]
            # Find the maximum value in the patch
            x_out[i, j] = np.max(patch)

    return x_out
```

```python
def duplicate(x_in):
    height, width = x_in.shape
    new_height = height * 2
    new_width = width * 2
    x_out = np.zeros((new_height, new_width))

    for i in range(height):
        for j in range(width):
            # Copy the pixel value to the corresponding 2x2 block in x_out
            x_out[2*i:2*i+2, 2*j:2*j+2] = x_in[i, j]

    return x_out
```

```python
# Now let's try max pooling back up
# The input x_high_res is the original high res image, from which you can deduce the position of the maximum index
def max_unpool(x_in, x_high_res):
  x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
  # TO DO -- write the subsampling routine
  # Replace this line
  x_out = x_out

  return x_out
```

```python
def max_unpool(x_in, x_high_res):
    height, width = x_high_res.shape
    new_height = height * 2
    new_width = width * 2
    x_out = np.zeros((new_height, new_width))

    for i in range(height):
        for j in range(width):
            if i < x_in.shape[0] and j < x_in.shape[1]:  # Ensure within bounds of x_in
                # Find the position of the max value in x_in corresponding to this 2x2 block
                max_index = np.unravel_index(np.argmax(x_in[i, j]), (2, 2))

                # Place the max value at the correct position in x_out
                x_out[2*i + max_index[0], 2*j + max_index[1]] = x_high_res[i, j]

    return x_out
```

Finally, we'll try upsampling using bilinear interpolation.  We'll treat the positions off the image as zeros by padding the original image and round fractional values upwards using np.ceil()

```python
def bilinear(x_in):
  x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
  x_in_pad = np.zeros((x_in.shape[0]+1, x_in.shape[1]+1))
  x_in_pad[0:x_in.shape[0],0:x_in.shape[1]] = x_in
  # TO DO -- write the duplication routine
  # Replace this line
  x_out = x_out

  return x_out
```

```python
def bilinear(x_in):
    height_in, width_in = x_in.shape
    height_out = height_in * 2
    width_out = width_in * 2

    x_out = np.zeros((height_out, width_out))

    # Pad the input image with zeros
    x_in_pad = np.pad(x_in, ((0, 1), (0, 1)), mode='constant')

    # Perform bilinear interpolation
    for i in range(height_out):
        for j in range(width_out):
            # Coordinates in the original image
            y_in = i / 2.0
            x_in = j / 2.0

            # Get the four neighboring pixels
            y_low = int(np.floor(y_in))
            y_high = min(y_low + 1, height_in)
            x_low = int(np.floor(x_in))
            x_high = min(x_low + 1, width_in)

            # Compute the weights for interpolation
            y_frac = y_in - y_low
            x_frac = x_in - x_low

            # Perform bilinear interpolation
            top_left = x_in_pad[y_low, x_low]
            top_right = x_in_pad[y_low, x_high]
            bottom_left = x_in_pad[y_high, x_low]
            bottom_right = x_in_pad[y_high, x_high]

            top_avg = top_left * (1 - x_frac) + top_right * x_frac
            bottom_avg = bottom_left * (1 - x_frac) + bottom_right * x_frac

            x_out[i, j] = top_avg * (1 - y_frac) + bottom_avg * y_frac

    return x_out
```
